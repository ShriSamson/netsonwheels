```python
# Function that extracts the category of the venue
def get_category_type(categories_list):
    return categories_list[0]["title"].encode('ascii',errors='ignore')

# Filter the category for each row
dataframeBusinesses['categories'] = dataframeBusinesses.apply(lambda row: get_category_type(row['categories']), axis=1)

dataframeBusinesses
```
